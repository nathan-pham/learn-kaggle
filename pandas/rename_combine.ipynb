# Rename and Combine

## Import Libraries

```python
import pandas as pd
```

## Import Dataset

```python
dataset = pd.read_csv("./datasets/chickweight.csv")
```

## Rename
`dataframe.rename()` allows you to rename indexes and columns  

```python
dataset.rename(columns=dict(zip(dataset.columns, dataset.columns.str.lower())))
```

```python
dataset.rename(index={ 0: "First Entry" })
```

Most of the time you"ll be renaming columns & rows rather than individual indexes

```python
dataset.rename_axis("chickens", axis="rows").rename_axis("features", axis="columns")
```

## Combine

```python
ca_videos = pd.read_csv("./datasets/videos/CAvideos.csv")
de_videos = pd.read_csv("./datasets/videos/DEvideos.csv")

pd.concat([ ca_videos, de_videos ]).head()
```

```python
left = ca_videos.set_index(["title", "trending_date"])
right = de_videos.set_index(["title", "trending_date"])

left.join(right, lsuffix="_CAN", rsuffix="_UK").head()
```
